```python
import pandas as pd
import plotly.express as px

from gtfparse import read_gtf

from datapaths import *

pd.set_option('display.max_rows', 4)
pd.set_option('display.max_columns', None)
```

```python
flipon_to_data = pd.read_table(F_FLIPON_TO_DATA)
flipon_to_data
```

```python
gtf = read_gtf(F_GENCODE)
gtf
```

```python
c_mir_list = ['c. M miR (+)', 'c. M miR (-)', 'c. MJ miR (+)', 'c. MJ miR (-)', 'c. J miR (+)', 'c. J miR (-)']
c_mir_list
```

```python
# All
with pd.option_context("display.max_rows", None):
    display(
        flipon_to_data[flipon_to_data["Gene Feature"] == "Promoter (<=1kb)"]
        .groupby("Gene ID")["Flipon"]
        .value_counts()
        .rename("Concurrent")
        .reset_index()[["Flipon", "Concurrent"]]
        .value_counts()
        .rename("Total")
        .reset_index()
        .pivot_table(values='Total', index='Flipon', columns='Concurrent').fillna('')
        .rename_axis(None, axis=0).rename_axis(None,axis=1)
    )
```

```python
# c. miRNA

with pd.option_context("display.max_rows", None):
    display(
        flipon_to_data[(flipon_to_data["Gene Feature"] == "Promoter (<=1kb)") & (flipon_to_data[c_mir_list].notna().sum(axis=1)>0)]
        .groupby("Gene ID")["Flipon"]
        .value_counts()
        .rename("Concurrent")
        .reset_index()[["Flipon", "Concurrent"]]
        .value_counts()
        .rename("Total")
        .reset_index()
        .pivot_table(values='Total', index='Flipon', columns='Concurrent').fillna('')
        .rename_axis(None, axis=1)
    )
```

```python
with pd.option_context('display.max_rows', 20):
    display(
        flipon_to_data[(flipon_to_data["Gene Feature"] == "Promoter (<=1kb)")]
        .groupby("Gene ID")[['Gene Name', 'Gene Type', "Flipon"]]
        .value_counts().rename('Conc').to_frame().sort_values('Conc', ascending=False).head(20)
    )
```

```python
table = (
    flipon_to_data[
        (flipon_to_data["Gene Feature"] == "Promoter (<=1kb)")
        & (flipon_to_data[c_mir_list].notna().sum(axis=1)>0)
    ]
    .groupby(["Gene ID", 'Transcript ID'])[["Gene Name", "Gene Type", "Flipon"]]
    .value_counts()
    .rename("Flipons per promoter")
    .to_frame()
    .sort_values("Flipons per promoter", ascending=False).reset_index()
)

table.to_csv(D_TABLES / "flipons_per_promoter_mir.tsv", sep='\t', index=False)

table
```
