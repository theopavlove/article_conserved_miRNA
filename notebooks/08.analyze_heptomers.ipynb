```python
import tempfile
import pandas as pd

from gtfparse import read_gtf
from Bio import SeqIO
from datapaths import *
from tqdm.notebook import tqdm

pd.set_option('display.max_rows', 4)
pd.set_option('display.max_columns', None)
```

```python
flipon_to_data = pd.read_table(F_FLIPON_TO_DATA)
flipon_to_data
```

```python
gtf = read_gtf(F_GENCODE)
gtf
```

```python
def get_heptomers(path_fa: str, name: str) -> pd.DataFrame:
    with tempfile.TemporaryDirectory() as tmpdir:
        jellyfish_bc = tmpdir + f"/jellyfish.bc"
        jellyfish_jf = tmpdir + f"/jellyfish.jf"
        jellyfish_fa = tmpdir + f"/jellyfish.fa"
        !jellyfish bc -m 7 -s 1G -t 60 -o {jellyfish_bc} {path_fa}
        !jellyfish count -m 7 -s 100M -t 60 --bc {jellyfish_bc} -o {jellyfish_jf} {path_fa}
        !jellyfish dump {jellyfish_jf} > {jellyfish_fa}

        fasta = list(SeqIO.parse(jellyfish_fa, "fasta"))
        heptomers, counts = [], []
        for rec in fasta:
            heptomers.append(str(rec.seq))
            counts.append(rec.description)

        df = (
            pd.DataFrame(data={"heptomer": heptomers, "heptomer_count": counts})
            .astype({"heptomer_count": int})
            .sort_values("heptomer_count", ascending=False)
            .reset_index(drop=True)
        )
        df['col_lvl_0'] = name.split('_')[0]
        df['col_lvl_1'] = name.split('_')[1]
        df['col_lvl_2'] = name.split('_')[2]
        df['total_regions'] = int(name.split('_')[3])
        df['total_heptomers'] = len(heptomers)
        return df
```

```python
def get_heptomers_for_flipons(flipon_type: str, mir: str):
    with tempfile.TemporaryDirectory() as tmpdir:
        # 1. Get flipon coordinates
        path_flipons = tmpdir + '/flipons.bed'
        path_flipons_fa = tmpdir + '/flipons.fa'

        flipons = flipon_to_data.copy()
        if mir == 'c. miR':
            flipons = flipons[(flipons.filter(like='c. ').notna().sum(axis=1) > 0)]
        if flipon_type!='Any flipon':
            flipons = flipons[flipons['Flipon'] == (flipon_type)]
        flipons = flipons['Coordinates'].str.split(r':|-', regex=True, expand=True)
        
        flipons.to_csv(path_flipons, sep='\t', header=False, index=False)
        !bedtools getfasta -bed {path_flipons} -fi {F_GENOME} > {path_flipons_fa}

        # 2. Get heptomer count
        heptomer_promoter_df = get_heptomers(path_flipons_fa, f"{flipon_type}_{mir}_Flipon_{flipons.shape[0]}")

        return heptomer_promoter_df


def get_heptomers_for_pp(transcript_id_list: list, name: str):
    with tempfile.TemporaryDirectory() as tmpdir:
        # 1. Get transcript coordinates        
        annotated_transcripts = gtf[(gtf['feature']=='transcript') & (gtf['transcript_id'].isin(transcript_id_list))].drop_duplicates('transcript_id')[['seqname', 'start', 'end', 'transcript_id', 'score', 'strand']]

        # 2. Get TSS coordinates
        annotated_transcripts['end'] = annotated_transcripts.apply(lambda row: row['start']+1 if row['strand']=='+' else row['end'], axis=1)
        annotated_transcripts['start'] = annotated_transcripts.apply(lambda row: row['end']-1 if row['strand']=='-' else row['start'], axis=1)

        path_tss = tmpdir + '/tss.bed'
        annotated_transcripts.to_csv(path_tss, sep='\t', header=False, index=False)

        # 3. Get promoter coordinates
        path_proximal_promoters = tmpdir + '/proximal_promoters.bed'
        path_proximal_promoters_fasta = path_proximal_promoters.replace('.bed', '.fa')
        !bedtools slop -i {path_tss} -g {F_CHROM_SIZES} -r 999 -l 1000 -s > {path_proximal_promoters}
        !bedtools getfasta -bed {path_proximal_promoters} -fi {F_GENOME} -s > {path_proximal_promoters_fasta}

        # 4. Get heptomer count
        heptomer_promoter_df = get_heptomers(path_proximal_promoters_fasta, f"{name}_{annotated_transcripts.shape[0]}")

        return heptomer_promoter_df
```

```python
transcript_id_list = gtf[gtf['feature']=='transcript']['transcript_id'].unique()
name = "All_All regions_PP"
df_all_all_pp = get_heptomers_for_pp(transcript_id_list, name)
df_all_all_pp['heptomer_global_rank'] = df_all_all_pp.index + 1
df_all_all_pp
```

```python
total_df = pd.DataFrame()

for x in tqdm(['Any flipon', 'g4', 'h-dna', 'sidd', 'z-dna']):
    for y in ['All regions', 'c. miR']:
        transcript_id_list = flipon_to_data.copy()
        if x != 'Any flipon':
            transcript_id_list = transcript_id_list[transcript_id_list['Flipon']==x]
        if y != 'All regions':
            transcript_id_list = transcript_id_list[transcript_id_list.filter(like='c. ').notna().sum(axis=1)>0]
        transcript_id_list = transcript_id_list['Transcript ID'].unique()
        name = f"{x}_{y}_PP"
        df_pp = get_heptomers_for_pp(transcript_id_list, name)
        
        df_flipon = get_heptomers_for_flipons(x, y)
        
        total_df = pd.concat([total_df, df_pp, df_flipon], ignore_index=True)

total_df = total_df.merge(df_all_all_pp[['heptomer', 'heptomer_global_rank']], on='heptomer', how='left')
total_df
```

```python
pivot = (
    pd.concat([df_all_all_pp, total_df], ignore_index=True)
    .pivot_table(index="heptomer", columns=["col_lvl_0", "col_lvl_1", "col_lvl_2"], values=["heptomer_count"], sort=False)
    .droplevel(0, axis=1)
)

pivot.to_csv(D_TABLES / 'heptomer_counts.tsv', sep='\t')
pivot
```
