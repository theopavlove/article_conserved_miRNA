```python
COLORS = {
    "g4": "229,50,55",
    "g4_empty": "255,175,175",
    "z-dna": "0,100,209",
    "z-dna_empty": "210,224,239",
    "sidd": "245,175,1",
    "sidd_empty": "225,226,154",
    "h-dna": "134,184,24",
    "h-dna_empty": "202,214,175",
}


def get_region_color(row: pd.Series) -> str:
    if row["region_name"] == "⠀":
        return COLORS[f"{row['Flipon']}_empty"]
    return COLORS[row["Flipon"]]


def get_region_name(row: pd.Series) -> str:
    mirna_p = ",".join(
        [f"{x.split(' ')[0]}(+)" for x in row["miRNA (+)"].split(", ")]
        if not row.isnull()["miRNA (+)"]
        else ""
    )
    mirna_m = ",".join(
        [f"{x.split(' ')[0]}(-)" for x in row["miRNA (-)"].split(", ")]
        if not row.isnull()["miRNA (-)"]
        else ""
    )
    res_name = ",".join((mirna_p, mirna_m)).strip(", ")
    if res_name == "":
        return "⠀"
    return res_name


TRACK_PARAMS = {
    "g4": f'track name="G4" description="Kouzine experimental G4 ({flipon_to_data[flipon_to_data["Flipon"]=="g4"].shape[0]:,d})" color={COLORS["g4"]} visibility=2 itemRgb="On"',
    "z-dna": f'track name="Z-DNA" description="Kouzine experimental Z-DNA ({flipon_to_data[flipon_to_data["Flipon"]=="z-dna"].shape[0]:,d})" color={COLORS["z-dna"]} visibility=2 itemRgb="On"',
    "sidd": f'track name="SIDD" description="Kouzine experimental SIDD ({flipon_to_data[flipon_to_data["Flipon"]=="sidd"].shape[0]:,d})" color={COLORS["sidd"]} visibility=2 itemRgb="On"',
    "h-dna": f'track name="H-DNA" description="Kouzine experimental H-DNA ({flipon_to_data[flipon_to_data["Flipon"]=="h-dna"].shape[0]:,d})" color={COLORS["h-dna"]} visibility=2 itemRgb="On"',
}

flipon_to_data_region_info = flipon_to_data.assign(
    region_name=lambda df: df.apply(get_region_name, axis=1),
    region_color=lambda df: df.apply(get_region_color, axis=1),
)
```

```python
for group_id, group_df in flipon_to_data_region_info.groupby("Flipon"):
    group_df[['chr', 'start', 'end']] = group_df['Coordinates'].str.split(r':|-', expand=True, regex=True)
    group_df['score'] = 0

    group_df = group_df[['chr', 'start', 'end', 'region_name', 'score', 'Strand', 'start', 'end', 'region_color']]
    
    path_out = D_GEBR / f"mm10.{group_id}.bed"
    group_df.to_csv(path_out, sep='\t', header=False, index=False)

    track_info = TRACK_PARAMS[group_id]
    !printf '%s\n%s\n' '{track_info}' "$(cat {path_out})" > {path_out}
```
