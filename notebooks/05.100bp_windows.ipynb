```python
import numpy as np
import pandas as pd
import tempfile

from gtfparse import read_gtf
from collections import Counter
from dataclasses import dataclass
from tqdm.notebook import tqdm

from datapaths import *

pd.set_option('display.max_rows', 4)
pd.set_option('display.max_columns', None)
```

```python
flipon_to_data = pd.read_table(F_FLIPON_TO_DATA)
flipon_to_data
```

```python
gtf = read_gtf(F_GENCODE)
gtf
```

```python
@dataclass
class Data:
    name: str
    path: str


flipon_data = (
    Data("g4", D_FLIPONS_BED / "mm10.actb_ssdna_enriched_g4.bed"),
    Data("z-dna", D_FLIPONS_BED / "mm10.actb_ssdna_enriched_z-dna.bed"),
    Data("sidd", D_FLIPONS_BED / "mm10.actb_ssdna_enriched_sidd.bed"),
    Data("h-dna", D_FLIPONS_BED / "mm10.actb_ssdna_enriched_h-dna.bed"),
)
```

```python
def get_overlaps_df(flipon_df: pd.DataFrame):
    overlaps_df: pd.DataFrame
    with tempfile.TemporaryDirectory() as tmpdir:
        # 1. Get transcript coordinates        
        transcript_list = flipon_df['Transcript ID'].unique()
        annotated_transcripts = gtf[(gtf['feature']=='transcript') & gtf['transcript_id'].isin(transcript_list)].drop_duplicates('transcript_id')[['seqname', 'start', 'end', 'transcript_id', 'score', 'strand']]

        # 2. Get TSS coordinates
        annotated_transcripts['end'] = annotated_transcripts.apply(lambda row: row['start']+1 if row['strand']=='+' else row['end'], axis=1)
        annotated_transcripts['start'] = annotated_transcripts.apply(lambda row: row['end']-1 if row['strand']=='-' else row['start'], axis=1)

        path_tss = tmpdir + '/tss.bed'
        annotated_transcripts.to_csv(path_tss, sep='\t', header=False, index=False)

        # 3. Get promoter coordinates
        path_proximal_promoters = tmpdir + '/proximal_promoters.bed'
        !bedtools slop -i {path_tss} -g {F_CHROM_SIZES} -r 999 -l 1000 -s > {path_proximal_promoters}

        # 4.1 Get 100 bp-windows
        path_windows = tmpdir + '/windows.bed'
        !bedtools makewindows -b {path_proximal_promoters} -w 100 -i srcwinnum > {path_windows}

        # 4.2. Reverse window numbers for + strand
        windows_df = pd.read_table(path_windows, header=None, names=['chr', 'start', 'end', 'window_name'])
        windows_df[['transcript_id', 'window_num']] = windows_df['window_name'].str.split('_', expand=True)
        windows_df = windows_df.merge(annotated_transcripts[['transcript_id', 'strand']].drop_duplicates(), how='left', on='transcript_id')

        new_windows = []
        for idx, group_df in windows_df.groupby('transcript_id', sort=False):
            if group_df.iloc[0,-1] == "+":
                group_df['window_num'] = [*range(-10, 0), *range(1, 11)]
            else:
                group_df['window_num'] = [*range(10, 0, -1), *range(-1, -11, -1)]
            new_windows.extend(group_df['window_num'].values)
        windows_df['window_num'] = new_windows
        windows_df['new_name'] = windows_df['transcript_id'] + '_' + windows_df['window_num'].astype(str)
        windows_df[['chr', 'start', 'end', 'new_name']].to_csv(path_windows, sep='\t', header=False, index=False)
        
        overlaps_df = pd.DataFrame()
        for flipon in flipon_data:
            path_to_intersection_count = tmpdir + f'/{flipon.name}_intersections.tsv'
            !bedtools intersect -a {path_windows} -b {flipon.path} -c > {path_to_intersection_count}

            print(flipon.name)
            !bedtools intersect -a {path_proximal_promoters} -b {flipon.path} -wo | wc -l
            
            df = pd.read_table(path_to_intersection_count, header=None, names=['Gene Chr', 'Gene Start', 'Gene End', 'Window Name', 'Count'])
            df['Flipon'] = flipon.name

            overlaps_df = pd.concat([overlaps_df, df], ignore_index=True)
        
        overlaps_df[['Gene Name', 'Window Num']] = overlaps_df['Window Name'].str.split('_', expand=True)

    return overlaps_df
```

```python
overlaps_df_all = get_overlaps_df(flipon_to_data)
overlaps_df_all
```

```python
overlaps_df_mirna = get_overlaps_df(flipon_to_data[flipon_to_data.filter(like='c. ').notna().sum(axis=1)>0])
overlaps_df_mirna
```

### Flipon distribution

```python
total_sum = np.zeros((4, 20))

for idx, group_df in tqdm(overlaps_df_all.groupby("Gene Name")):
    pivot_df = group_df.astype({"Window Num": int}).pivot_table(
        values=["Count"], index=["Flipon"], columns=["Window Num"], sort=True
    )
    total_sum += pivot_df.values

total_sum_df = pd.DataFrame(
    total_sum, index=["g4", "h-dna", "sidd", "z-dna"], columns=[*range(-10, 0), *range(1, 11)], dtype=int
)

total_sum_df.to_csv(D_TABLES / "total_window_flipons_all.tsv", sep='\t', index=True)
display(total_sum_df)
```

### Pairwise sums

```python
flipon_type_num = {"g4": 1, "h-dna": 2, "sidd": 3, "z-dna": 4}
flipon_num_type = {"1": "G", "2": "H", "3": "S", "4": "Z"}
# flipon_num_type = {str(val):key for key,val in flipon_type_num.items()}
counter = Counter()

for idx, group_df in tqdm(overlaps_df_all.groupby("Gene Name")):
    group_df["Count"] = group_df["Count"].apply(lambda x: int(x != 0))
    group_df["Count_type"] = group_df[["Flipon", "Count"]].apply(
        lambda x: flipon_type_num[x["Flipon"]] if x["Count"] != 0 else 0, axis=1
    )

    pivot_df = (
        group_df.astype({"Window Num": int})
        .pivot_table(
            values=["Count_type"], index=["Flipon"], columns=["Window Num"], sort=True
        )
        .astype(str)
    )

    pivot_sum = pivot_df.sum().str.replace("0", "")
    for col, val in zip(pivot_sum.index, pivot_sum.values):
        if val == "":
            continue
        counter[(col[1], val)] += 1

result = pd.DataFrame()
for idx, val in counter.items():
    result.loc[idx[1], idx[0]] = val

result = result.fillna('')
result = result.loc[:,sorted(result.columns)]
# result.columns = [f"[{c*100}, {(c-1)*100}]" for c in result.columns]
result = result.sort_index()
result.index = [' & '.join([flipon_num_type[y] for y in x]) for x in result.index]

with pd.option_context('display.max_rows', None):
    display(result)

```

### Summary for all flipons

```python
counter = Counter()

for idx, group_df in tqdm(overlaps_df_all.groupby("Gene Name")):
    group_df["Count"] = group_df["Count"].apply(lambda x: int(x != 0))

    pivot_df = group_df.pivot_table(
        values=["Count"], index=["Flipon"], columns=["Window Num"]
    )

    pivot_sum = pivot_df.sum()
    for col, val in zip(pivot_sum.index, pivot_sum.values):
        counter[(int(col[1]), val)] += 1

result = pd.DataFrame()
for idx, val in counter.items():
    result.loc[idx[1], idx[0]] = val
result = result.loc[:,sorted(result.columns)].sort_index().fillna('')

with pd.option_context('display.max_rows', None):
    display(result)
```
